```python
import sklearn.model_selection
x_ucz, x_test =\
sklearn.model_selection.train_test_split(np.arange(x.shape[0]), 
                                        test_size=0.2,
                                        random_state=1234)
```

```python
X_ucz, X_test = x.iloc[x_ucz, :], x.iloc[x_test, :]
y_ucz, y_test = y[x_ucz], y[x_test]
X_ucz.shape, X_test.shape, y_ucz.shape, y_test.shape
```

```python
y_test[y_test==1].shape[0]
```

```python
def fit_classifier(alg, X_ucz, X_test, y_ucz, y_test):
      
    alg.fit(X_ucz, y_ucz)
    y_pred_ucz = alg.predict(X_ucz)
    y_pred_test = alg.predict(X_test)
    return {
        "ACC_ucz": sklearn.metrics.accuracy_score(y_ucz, y_pred_ucz),
        "ACC_test": sklearn.metrics.accuracy_score(y_test, y_pred_test),
        "P_ucz":   sklearn.metrics.precision_score(y_ucz, y_pred_ucz),
        "P_test":   sklearn.metrics.precision_score(y_test, y_pred_test),
        "R_ucz":   sklearn.metrics.recall_score(y_ucz, y_pred_ucz),
        "R_test":   sklearn.metrics.recall_score(y_test, y_pred_test),
        "F1_ucz":  sklearn.metrics.f1_score(y_ucz, y_pred_ucz),
        "F1_test":  sklearn.metrics.f1_score(y_test, y_pred_test)
    }
```

```python
from sklearn import tree
```

```python
params = ["tree 20"]
res = [fit_classifier(sklearn.tree.DecisionTreeClassifier(max_depth=20,criterion='entropy'),
                      X_ucz, X_test, y_ucz, y_test)]
df_results = pd.DataFrame(res, index=params)
```

```python
df_results
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
params.append("forest 20")
res.append(fit_classifier(sklearn.ensemble.RandomForestClassifier(max_depth=20,criterion='entropy'),
                          X_ucz, X_test, y_ucz, y_test))
df_results = pd.DataFrame(res, index=params)
```

```python
df_results
```

```python
import sklearn.neighbors
```

```python
params.append("knn_std")
res.append(fit_classifier(sklearn.neighbors.KNeighborsClassifier(),
                          X_ucz, X_test, y_ucz, y_test))
df_results = pd.DataFrame(res, index=params)
```

```python
df_results
```

```python
from sklearn.neural_network import MLPClassifier
```

```python
params.append("nn")
res.append(fit_classifier(sklearn.neural_network.MLPClassifier(),
                          X_ucz, X_test, y_ucz, y_test))
df_results = pd.DataFrame(res, index=params)
```

```python
df_results
```

```python
import numpy as np
import pandas as pd
from sklearn import tree
from sklearn.model_selection import GridSearchCV



parameters = {'max_depth':range(20,25), 'max_leaf_nodes':range(110,140), 'max_features':range(1,15), 'min_samples_leaf':range(1,4)}
clf = GridSearchCV(tree.DecisionTreeClassifier(), parameters, n_jobs=4,scoring='f1')
clf.fit(X=x, y=y)
tree_model = clf.best_estimator_
print (clf.best_score_, clf.best_params_) 
```

```python
parameters = {'max_depth':range(1,20)}
clf = GridSearchCV(sklearn.ensemble.RandomForestClassifier(), parameters, n_jobs=3,scoring='f1')
clf.fit(X=x, y=y)
tree_model = clf.best_estimator_
print (clf.best_score_, clf.best_params_) 
```

```python
parameters = {'n_neighbors':range(1,10), 'leaf_size':range(1,5)}
clf = GridSearchCV(sklearn.neighbors.KNeighborsClassifier(), parameters, n_jobs=3,scoring='f1')
clf.fit(X=x, y=y)
tree_model = clf.best_estimator_
print (clf.best_score_, clf.best_params_) 
```
